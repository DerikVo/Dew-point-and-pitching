# Correlation plots

Based on our correlation matrix I wanted to plot the features (that we selected) and get some statistical information such as R values and RMSE. The concept was taken from [Nick Wan's datas cience boot camp](https://www.youtube.com/watch?v=Du7y0ChihKE).

```python
def cor_plot(feat1, feat2, _df):
    """
    Creates a Joint plot of two features and prints out the statistics such as Pearson R, SPearmen R, T-Test, and RMSE.
    Concept taken from Nick Wan's data science bootcamp on youtube.
    =============================================================================    
    Keyword arguments:
    Feat 1 -- The first feature you're comparing
    Feat 2 -- The Second feature you're comparing
    _df -- The dataframe youre using
    
    =================================================
    Example:
    cor_plot('INDUCED_VERTICAL_BREAK', 'VERTICAL_APPROACH_ANGLE', model_df)
    
    """
    r= pearsonr(_df[feat1], df[feat2])
    r2 = r[0]**2
    
    sr = spearmanr(_df[feat1], df[feat2])
    sr2= sr[0]**2
    sr[0], sr2, sr[-1]
    
    t= ttest_ind(_df[feat1], df[feat2])
    t[0], t[-1]
    
    #rmse
    rmse= mean_squared_error(_df[feat1], df[feat2])**.5
    title= f"""
    {feat1} and {feat2}
    Pearson r: {round(r[0],3)} P-value: {round(r[-1],3)}
    Spearmen r: {round(sr[0],3)} P-value: {round(sr[-1],3)}
    T-test: {round(t[0],3)} P-value: {round(t[1],3)}
    RMSE: {round(rmse,3)}
    """
    sns.jointplot(data=_df, x=feat1,y =feat2, kind='reg')
    plt.title(title)
    plt.show()
    print(title)
```

```python
#performing a t-test to see if the environmental factors had an effect on the pitch
t_stat, p_value = ttest_1samp(model_df['DIFFERENCE'], 0)
print(t_stat, p_value)
```

Given our P value, we can reject the null hypothesis and conclude that the environmental factors from horizontal break is statistically significant. # Modeling

Here we will build models to predict the drop in approach angel when accounting for atmospheric factors and when not accounting for the atmospheric factors. I want to do it with a logistic regression and use predict_proba then find the difference between using different target variables; however, out target variable is continuous so ill have to use a linear regression. 

```python
#setting the PID as the index so its not included in the model, and be keep the PID
model_df.set_index(['PID'], inplace=True)
```

```python
model_df.head(5)
```

```python
#although a negative value (from the horiszonal break) indicates if it was thrown via left or right, we will do greater than zero for testing purposes
model_df['is_dew'] = model_df['DIFFERENCE'].apply(lambda x: 1 if x > 0 else 0)
```

```python
model_df.head(10)
```

```python
#model including horizontal break
X= model_df[['RELEASE_SPEED','SPIN_RATE_ABSOLUTE', 'VERTICAL_APPROACH_ANGLE', 'RELEASE_HEIGHT','HORIZONTAL_BREAK']]
y = model_df['is_dew']
X1_train, X1_test, y1_train, y1_test = train_test_split(X, y, random_state = 42)
lr1 = LogisticRegression()
lr_angle= lr1.fit(X1_train, y1_train)
```

```python
lr_angle.predict_proba(X1_train)[:,1]
```

```python
#model without horizontal break
X= model_df[['RELEASE_SPEED','SPIN_RATE_ABSOLUTE', 'VERTICAL_APPROACH_ANGLE', 'RELEASE_HEIGHT','HORIZONTAL_APPROACH_ANGLE']]
y = model_df['is_dew']
X2_train, X2_test, y2_train, y2_test = train_test_split(X, y, random_state = 42)
lr2 = LogisticRegression()
lr_break= lr2.fit(X2_train, y2_train)
```

```python
lr_break.predict_proba(X2_train)[:,1]
```
